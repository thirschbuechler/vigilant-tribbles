```python
import numpy as np
import matplotlib.pyplot as plt


rng=np.random.RandomState(0)

# shape
xsets = 20
ysets = 10

# test data
data=rng.randn(xsets,ysets)

# xaxis linspace
xmax = xsets
ymax = ysets

# square pixels
aspect = 0.5 * xmax / ymax
# nonsquare pixels
# aspect = "auto"

# axes xticks xticklabels
extent = np.array([0, ymax, 0, xmax])

# scale factor for pixels
#   derived from fontsize 22
#   division over 72: pt to inch
#   scaled with one

fa = 22 / 72 * 1.0
xfig = xsets * fa
yfig = ysets * fa

plt.figure(figsize=(xfig,yfig))

plt.imshow(data, origin = 'lower',  extent = extent, aspect = aspect);
```

```python
import numpy as np
import matplotlib.pyplot as plt


rng=np.random.RandomState(0)

# shape
xsets = 20
ysets = 10

# test data
data=rng.randn(xsets,ysets)

# xaxis linspace
xmax = xsets
ymax = ysets

# square pixels
aspect = 0.5 * xmax / ymax
# nonsquare pixels
# aspect = "auto"

# axes xticks xticklabels
#extent = np.array([0, ymax, 0, xmax])
#extent = None
extent = np.array([500, 1000*ymax, 0, xmax])

# scale factor for pixels
#   derived from fontsize 22
#   division over 72: pt to inch
#   scaled with one

fa = 22 / 72 * 1.0

# correct for different x or y axis scaling
aspect = aspect * ((extent[1]-extent[0])/(extent[3]-extent[2]))*2

xfig = xsets * fa
yfig = ysets * fa

plt.figure(figsize=(xfig,yfig))

plt.imshow(data, origin = 'lower',  extent = extent, aspect = aspect);
```
